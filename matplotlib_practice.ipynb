```python
#add this as first line' "%" enables to use back-end processor and inline enables to view the graph immediately next to the cell when invoked.
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Plot the graph

plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#line chart using object oriented interface

#create a figure
fig, ax = plt.subplots()
#define axes, inside parantheses axes can be passed separated by a comma.
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#quick plot using OO interface

ax = plt.axes()
#Bax.plot(x_axis, y_axis)
```

```python
#Bar chart example

plt.bar(x_axis, y_axis, color = "Green", linewidth = 2, label = "Bar Chart")
#create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#create range for y axis
plt.ylim(0,45)
#create the title for the chart
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Horizontal Bar chart example

plt.barh(x_axis, y_axis, color = "Green", label = "Bar Chart")
#invert y axis to get it in right order; use gca method

plt.gca().invert_yaxis()

#create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#create range for y axis
plt.xlim(0,45)
#create the title for the chart
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#bar chart using object oriented method

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "black", label = "New York")
plt.legend()
```

```python
#create horizontal bar chart
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python

# to make the pie chart more readable, adjust size, use figsize

plt.subplots(figsize=(10,10))

#to show the seventh element - the largest size separately, use explode
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)

#to show the percentage within the pie use autopct
plt.pie(y_axis, explode=explode_values, labels = x_axis, autopct="%.1f%%")
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
